```python
import numpy as np
import pandas as pd
mylist = list('abcdefghjklmnopqrstuvwxyz')
df_list = pd.DataFrame(mylist,columns = ['list'] )
myarr = np.arange(26) 
df_array = pd.DataFrame(myarr, columns = ['arr'])
mydict = dict(zip(mylist, myarr))
df_dict = pd.Series(mydict)
print("DataFrame từ List:")
print(df_list)

print("\nDataFrame từ Numpy Array:")
print(df_array)

print("\nDataFrame từ Dictionary:")
print(df_dict)
```
